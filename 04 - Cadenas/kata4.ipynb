```python
text = """Interesting facts about the Moon. The Moon is Earth's only satellite. There are several interesting facts about the Moon and how it affects life here on Earth. On average, the Moon moves 4cm away from the Earth every year. This yearly drift is not significant enough to cause immediate effects on Earth. The highest daylight temperature of the Moon is 127 C."""

print(text.split(". "))
```

Después de tener los datos divididos, se van a especificar unas palabras clave para que se puedan comparar en una oración.

```python
words = ["average", "temperature", "distance"]
```

Posteriormente se realizará un loop para que se impriman las oraciones que contengan las palabras clave.

```python
textS = text.split(". ")

words = ["average", "temperature", "distance"]

for sentences in textS:
    for keyW in words:
        if keyW in sentences:
            print(sentences)
            break
```

Después se reemplazará el caracter "C" a Celcius

```python
text = """Interesting facts about the Moon. The Moon is Earth's only satellite. There are several interesting facts about the Moon and how it affects life here on Earth. On average, the Moon moves 4cm away from the Earth every year. This yearly drift is not significant enough to cause immediate effects on Earth. The highest daylight temperature of the Moon is 127 C."""

textS = text.split(". ")

words = ["average", "temperature", "distance"]

for sentences in textS:
    for keyW in words:
        if keyW in sentences:
            print(sentences.replace("C", "Celcius"))
            break
```

# Formato

Usando las distintas formas de aplicarlo

```python
satellite = "Ganymede"
gravity = 0.00143
planet = "Mars"
title = f"gravity on the satellite {satellite}".title()
info = f"-------------------------\nPlanet's name: {planet}\nGravity on the satellite {satellite}: {gravity * 1000} m/s2"

template = f"{title}\n{info}"

print(template)
```

```python
satellite = "Ganymede"
gravity = 0.00143
planet = "Mars"

template = """
Gravity on the Satellite: {satellite}
------------------------------------------------
Planet's name: {planet}
Gravity on the satellite {satellite}: {gravity} m/s2
"""
print(template.format(satellite = satellite, planet = planet, gravity = gravity*1000))
```
