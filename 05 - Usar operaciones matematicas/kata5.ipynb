#  Crear un programa que calcule la distancia entre dos planetas

```python
planet1 = 149597870
planet2 = 778547200

distance = abs(planet1 - planet2)

print(f"The distance is {distance} km")
print(f"The distance is {distance*0.621} miles")
```
