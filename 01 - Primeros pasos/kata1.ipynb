# Creación de programa para dar la fecha

```python
from datetime import date

print("Today's date is: " + str(date.today()))
```

Se usó la librería correspondiente y se convirtió el dato de fecha a string para que pueda dser concatenado con el texto previo.

## Creación de un programa para convertir entre parsecs y años luz

```python
parsec = input("How many parsecs are?")

lightyears = int(parsec) * 3.26156

print(str(parsec) + " parsec, is " + str(lightyears) + " lightyears")
```

Se implementó un input para que el usuario pueda convertir los parsec que desee, este dato fue convertido a int para que pueda ser multiplicado por los años luz a los que equivale, posteriormente esta información se transformó en str para que pueda ser concatenada con elementos de la misma naturaleza.
