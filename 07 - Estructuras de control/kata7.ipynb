# Ciclos While y For

### Lo siguiente es un programa que acepta varias entradas de texto y deja de recibirlas al escribir una palabra clave.

```python
newPlanet = ""
planets = []

while newPlanet != "done":
    if newPlanet:
        planets.append(newPlanet)
    newPlanet = input("Name a planet, if finished, type Done").lower()

for planet in planets:
    print(planet.capitalize())
print("This is what you wrote.")

```
