## Trabajando con datos en una lista

### Se realizar치 un programa que brinde los planetas m치s cercanos y m치s lejanos al sol a partir del planeta elegido

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
planetChosen = input("Choose a planet on the solar system")
planetChoseni = planets.index(planetChosen)

planetsClose = planets[0:planetChoseni]
planetsFar = planets[planetChoseni+1:]

print(f"The closest planets to the sun are {', '.join(planetsClose)}")
print(f"The furthest planets to the sun are {', '.join(planetsFar)}")
```
