# Crear un programa que advierta de un asteroide según su velocidad

```python
velAsteroide = 20

if velAsteroide > 25:
    print("WARNING: AN ASTEROID IS COMING TO EARTH.")
else:
    print("Low activity of asteroids")
```

## Crear un programa que indique que será visible un asteroide desde la tierra.

```python
velAsteroide = 19

if velAsteroide >= 20:
    print("A beam of light will be seen")
else:
    print("Clear skies")
```

## Crear un programa que brinde información completa a las personas

```python
tamAsteroide = 30
velAsteroide = 50

if tamAsteroide > 25 and velAsteroide > 25:
    print("WARNING: AN ASTEROID IS COMING TO EARTH, SEVERAL DAMAGE WILL BE CAUSED")
elif tamAsteroide <= 25 and velAsteroide >= 20:
    print("A beam of light will be seen")
else:
    print("Low activity of asteroids")
```
