# Funciones

### Lo siguiente brinda un informe de cuanto combustible hay en los tanques brindados, así como su promedio

```python
def text(tank1, tank2, tank3):
    average = (tank1 + tank2 + tank3) / 3
    return (f"""
    The tank no. 1 has {tank1} liters.
    The tank no. 2 has {tank2} liters.
    The tank no. 3 has {tank3} liters.
    The average of the tanks is {average} liters.
    """)

print(text(10, 0, 5))
```

### Se actualiza la función para que el promedio se haga por separado

```python
def average(tanks):
    add = sum(tanks)
    average = add / 3
    return average

def text(tank1, tank2, tank3):
    return(f"""
    The tank no. 1 has {tank1} liters.
    The tank no. 2 has {tank2} liters.
    The tank no. 3 has {tank3} liters.
    The average of the tanks is {average([tank1, tank2, tank3])} litres.
    """)

print(text(5, 20, 20))
```

# Argumentos de palabra clave

### Lo siguiente es un programa que brinda informes llamando a una función según la información brindada

```python
def text(preF, timeF, arrival, fLeft, fRight):
    return(f"""
    The destination is: {arrival}.
    The flight time is: {preF + timeF} minutes.
    The fuel left is: {fLeft + fRight} litres.
    """)

print(text(14, 51, "Moon", 200000, 300000))
```

### A continuación se modifica el código para que los datos de tiempo y de combustible no sean limitados

```python
def text(arrival, *timeF, **fuel):
    return(f"""
    The destination is: {arrival}.
    The flight time is: {sum(timeF)} minutes.
    The fuel left is: {sum(fuel.values())} litres.
    """)

print(text("Moon", 10, 10, left = 100, right = 100))
```

### Por último se realiza una modificación más para que los tanques se diferencíen entre sí con sus respectivos nombres y cantidades

```python
def text(arrival, *timeF, **fuel):
    textWhole = (f"""
    The destination is: {arrival}.
    The flight time is: {sum(timeF)} minutes.
    """)
    for tank, amount in fuel.items():
        textWhole = textWhole + (f"The {tank} tank hast {amount} litres left.\n    ")
    return textWhole

print(text("Moon", 10, 10, left = 100, right = 100, middle = 100))
```
